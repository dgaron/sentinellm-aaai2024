```python
%%bash
models="xlnet-base-cased sentence-transformers/stsb-bert-base"
keys="llm123 nlp2023"
embedding_rescaling_factors="3 10"

for model in $models: do
  for key in $keys; do
    for erf in $embedding_rescaling_factors; do
      MODEL_PATH="${model}-blake2-glide${erf}-key-${key}-shuffled"
      echo $MODEL_PATH
      python llm_converter.py \
        --model_name_or_path $model \
        --encryption_key $key \
        --destination $MODEL_PATH \
        --embedding_rescaling_factor $erf \
        --shuffle
    done
  done
done
```
